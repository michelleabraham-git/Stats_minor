# Functions and if statements

```python
# to write functions we use def command , # putting a colon means the function starts from here
# for multiline comments we use 3 quotes, for single line comments wwe use hash (#)

def has_student_passed(score) :
  '''
  This is a multiline comment.
  This function takes the score and returns pass if
  the score if above 40.
  '''
  if score >= 40:
    print("the student has passed") # executes if, if condition evaluates to true
  else:
    print("the student has failed") # executes if, if condition evaluate to false
  return
```

# For loop

```python
for s in score:
  print (f"the score is {s}")
  has_student_passed(s)
```

```python
# you can also do it this way
for s in score:
  print (f"the score is {s}", {has_student_passed(s)}) # if you are putting anything in the f string that has to give a value and not just print as string you have to put the curly bracket {}
  # the none showing is the return value, if u put some value to return like 0 it'll show 0 and not none
```

#Homework :
*   Use this dictionary along with the for loop to print the name and the score of the student
*   You can also try calculating the grade of each student inside the for block






```python
## Homework answer:
names = ["Alex", "Bob", "Charlie", "Emma"]
scores = [65, 78, 90, 30]

scores_dict = dict(zip(names, scores))
def get_grade(score):
    if score < 40:
        return "F"
    elif score <= 50:
        return "E"
    elif score <= 60:
        return "D"
    elif score <= 70:
        return "C"
    elif score <= 80:
        return "B"
    else:
        return "A"
for name, score in scores_dict.items():
    grade = get_grade(score)
    print(f"Name: {name}, Score: {score}, Grade: {grade}")
```
