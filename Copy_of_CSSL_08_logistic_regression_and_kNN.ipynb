```python
# Import necessary packages
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
iris = load_iris()
X = iris.data[:, :4]
y = iris.target
type(X)
```

One hot encoder is used when we do not want to impose an order onto the variables

Here, y is already encoded. We do not need to use LabelEncoder

```python
scaled = StandardScaler()
X_transformed = scaled.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.3, random_state=0)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
print(accuracy_score(y_train, y_pred_train))
print(accuracy_score(y_test, y_pred_test))
#help(accuracy_score)
```

```python
from sklearn.linear_model import LogisticRegression
X_train_2d = X_train[:, :2]

model_2d = LogisticRegression()
model_2d.fit(X_train_2d, y_train)

display = DecisionBoundaryDisplay.from_estimator(
    model_2d, X_train_2d, plot_method="contourf", response_method="predict",
    xlabel="Feature 1", ylabel="Feature 2", grid_resolution=200
)

plt.scatter(X_train_2d[:, 0], X_train_2d[:, 1], c=y_train, edgecolors="k", s=20)
plt.title("Logistic Regression Decision Boundary (First 2 Features)")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.neighbors import KNeighborsClassifier

X_train_2d_knn = X_train[:, :2]

knn_model_2d = KNeighborsClassifier(n_neighbors=5)
knn_model_2d.fit(X_train_2d_knn, y_train)
display_knn_contour = DecisionBoundaryDisplay.from_estimator(
    knn_model_2d, X_train_2d_knn, plot_method="contour", response_method="predict",
    xlabel="Feature 1", ylabel="Feature 2", grid_resolution=200, eps=0.5
)

plt.scatter(X_train_2d_knn[:, 0], X_train_2d_knn[:, 1], c=y_train, edgecolors="k", s=20)
plt.title("kNN Decision Boundary (First 2 Features) with Contour")
plt.show()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm_logistic = ConfusionMatrixDisplay.from_predictions(y_test, y_pred_test, cmap=plt.cm.Blues)
plt.title('Confusion Matrix for Logistic Regression (Test Data)')
plt.show()
```
